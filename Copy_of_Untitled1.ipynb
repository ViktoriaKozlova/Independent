```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import pytesseract
import cv2
```

```python
image = cv2.imread('//ex.png')
```

```python
string = pytesseract.image_to_string(image)
print(string)
```

```python
image_copy = image.copy()
target_word = "the"
data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)
```

```python
word_occurences = [ i for i, word in enumerate(data["text"]) if word.lower() == target_word ]
```

```python
for occ in word_occurences:
    w = data["width"][occ]
    h = data["height"][occ]
    l = data["left"][occ]
    t = data["top"][occ]
    p1 = (l, t)
    p2 = (l + w, t)
    p3 = (l + w, t + h)
    p4 = (l, t + h)
    image_copy = cv2.line(image_copy, p1, p2, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p2, p3, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p3, p4, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p4, p1, color=(255, 0, 0), thickness=2)
```

```python
plt.imsave("all_words.png", image_copy)
plt.imshow(image_copy)
plt.show()
```

```python
# чтобы нарисовать сделаем копию изображения
image_copy = image.copy()
# слово для поиска
target_word = "robots"
# получить все данные из изображения
data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

target_word1 = "of"
# получить все данные из изображения
data1 = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)
```

```python
word_occurences = [ i for i, word in enumerate(data["text"]) if word.lower() == target_word ]
word_occurences1 = [ i for i, word in enumerate(data["text"]) if word.lower() == target_word1 ]
```

```python
for occ in word_occurences:
    # извлекаем ширину, высоту, верхнюю и левую позицию для обнаруженного слова
    w = data["width"][occ]
    h = data["height"][occ]
    l = data["left"][occ]
    t = data["top"][occ]

    # определяем все точки окружающей рамки
    p1 = (l, t)
    p2 = (l + w, t)
    p3 = (l + w, t + h)
    p4 = (l, t + h)
    # рисуем 4 линии (прямоугольник)
    image_copy = cv2.line(image_copy, p1, p2, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p2, p3, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p3, p4, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p4, p1, color=(255, 0, 0), thickness=2)
```

```python
for occ in word_occurences1:
    w1 = data1["width"][occ]
    h1 = data1["height"][occ]
    l1 = data1["left"][occ]
    t1 = data1["top"][occ]
    # определяем все точки окружающей рамки

    p11 = (l1, t1)
    p21 = (l1 + w1, t1)
    p31 = (l1 + w1, t1 + h1)
    p41 = (l1, t1 + h1)
    # рисуем 4 линии (прямо
    image_copy = cv2.line(image_copy, p11, p21, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p21, p31, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p31, p41, color=(255, 0, 0), thickness=2)
    image_copy = cv2.line(image_copy, p41, p11, color=(255, 0, 0), thickness=2)
```

```python
plt.imsave("all_words.png", image_copy)
plt.imshow(image_copy)
plt.show()
```
